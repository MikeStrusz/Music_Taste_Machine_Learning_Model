```python
def get_albums(week: str):
    files = get_all_prediction_files()
    target_file = None
    for file, _, formatted in files:
        if week == formatted or week == file:
            target_file = file
            break
    if not target_file:
        raise HTTPException(status_code=404, detail="Week not found")
    records, analysis_date = load_predictions(target_file)
    # Attach gut scores if any
    gut_scores = load_master_gut_scores()
    gut_dict = { (g['Artist'], g['Album']): g for g in gut_scores if 'Artist' in g and 'Album' in g }
    for r in records:
        key = (r.get('Artist'), r.get('Album'))
        if key in gut_dict:
            r['gut_score'] = gut_dict[key]['gut_score']
            r['notes'] = gut_dict[key].get('notes', '')
    return {"week": analysis_date, "albums": records}
```

```python
import pandas as pd
import os
import glob
from datetime import datetime

# Simulate what the API does
week = "March 20, 2026"

# Find the file
prediction_files = []
for f in glob.glob('predictions/*_Album_Recommendations.csv'):
    date_str = os.path.basename(f).split('_')[0]
    try:
        date_obj = datetime.strptime(date_str, '%m-%d-%y')
        if date_obj >= datetime(2025, 2, 1):
            formatted = date_obj.strftime('%B %d, %Y')
            prediction_files.append((f, date_obj, formatted))
    except:
        continue
prediction_files.sort(key=lambda x: x[1], reverse=True)
target_file = None
for file, _, formatted in prediction_files:
    if week == formatted:
        target_file = file
        break
if not target_file:
    print("Week not found")
else:
    print("Target file:", target_file)

    # Load predictions
    df = pd.read_csv(target_file)
    # Normalize columns
    if 'Album' in df.columns and 'Album Name' not in df.columns:
        df['Album Name'] = df['Album']
    if 'Artist Name(s)' in df.columns and 'Artist' not in df.columns:
        df['Artist'] = df['Artist Name(s)']
    records = df.to_dict(orient='records')

    # Load covers
    covers_file = 'data/nmf_album_covers.csv'
    if os.path.exists(covers_file):
        covers_df = pd.read_csv(covers_file)
        print("Covers columns:", covers_df.columns.tolist())
        # Normalize covers columns
        if 'Artist Name(s)' in covers_df.columns:
            covers_df = covers_df.rename(columns={'Artist Name(s)': 'Artist'})
        if 'Album Name' in covers_df.columns and 'Album' not in covers_df.columns:
            covers_df = covers_df.rename(columns={'Album Name': 'Album'})
        print("After rename:", covers_df.columns.tolist())
        print("First few covers:\n", covers_df.head())
    else:
        print("No covers file")
        covers_df = pd.DataFrame()

    # Merge (just first 3 albums)
    for r in records[:3]:
        artist = r.get('Artist')
        album = r.get('Album')
        print(f"Checking {artist} - {album}")
        if covers_df.empty:
            r['Album Art'] = None
        else:
            cover_match = covers_df[(covers_df['Artist'] == artist) & (covers_df['Album'] == album)]
            if not cover_match.empty:
                r['Album Art'] = cover_match.iloc[0]['Album Art']
            else:
                r['Album Art'] = None
        print(f"  -> Album Art: {r['Album Art']}")

    print("First album:", records[0]['Artist'], records[0]['Album'], records[0]['Album Art'])
```

```python
import networkx as nx
import pandas as pd
import os

# Build graph the same way as in api.py
def build_graph():
    training_file = 'data/2026_training_complete_with_features.csv'
    liked_artists = set()
    if os.path.exists(training_file):
        df = pd.read_csv(training_file)
        liked_mask = df['source_type'].isin(['top_100_ranked', 'honorable_mention'])
        liked_artists.update(df[liked_mask]['Artist Name(s)'].dropna().str.split(',').explode().str.strip())

    similar_df = pd.DataFrame()
    similar_file = 'data/liked_artists_only_similar.csv'
    if os.path.exists(similar_file):
        similar_df = pd.read_csv(similar_file)

    G = nx.Graph()
    for artist in liked_artists:
        G.add_node(artist, type='liked')

    for _, row in similar_df.iterrows():
        artist = row.get('Artist', '')
        if not artist:
            continue
        similar_str = row.get('Similar Artists', '')
        if pd.isna(similar_str):
            continue
        similar_list = [s.strip() for s in similar_str.split(',')]
        G.add_node(artist, type='liked_similar')
        for s in similar_list:
            G.add_node(s, type='similar')
            G.add_edge(artist, s)

    return G

G = build_graph()
print("Lucy Dacus" in G)
print("Phoebe Bridgers" in G)
print("Samia" in G)
```

```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time

vault = pd.read_csv('data/vault_clean.csv')

client_id = "71faef9605da4db495b691d96a0daa4b"
client_secret = "832e40da22e049bba93f29d9dbeb2e62"

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=client_id,
    client_secret=client_secret
))

results = []
for i, row in vault.iterrows():
    artist = row['Artist']
    album = row['Album']
    try:
        query = f"artist:{artist} album:{album}"
        res = sp.search(q=query, type='album', limit=1)
        items = res['albums']['items']
        if items:
            url = items[0]['external_urls']['spotify']
            results.append({'Artist': artist, 'Album': album, 'Spotify URL': url})
            print(f"✅ {artist} — {album}")
        else:
            # Fallback: looser search
            query2 = f"{artist} {album}"
            res2 = sp.search(q=query2, type='album', limit=1)
            items2 = res2['albums']['items']
            if items2:
                url = items2[0]['external_urls']['spotify']
                results.append({'Artist': artist, 'Album': album, 'Spotify URL': url})
                print(f"🟡 {artist} — {album} (loose match)")
            else:
                results.append({'Artist': artist, 'Album': album, 'Spotify URL': None})
                print(f"❌ {artist} — {album}")
    except Exception as e:
        print(f"⚠️ {artist} — {album}: {e}")
        results.append({'Artist': artist, 'Album': album, 'Spotify URL': None})
    time.sleep(0.1)  # be nice to the API

links_df = pd.DataFrame(results)
links_df.to_csv('data/vault_spotify_links.csv', index=False)
found = links_df['Spotify URL'].notna().sum()
print(f"\nDone! Found {found} / {len(vault)} links")
```
