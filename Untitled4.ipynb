```python
import pandas as pd, glob

genres = set()
for file in glob.glob('predictions/*_Album_Recommendations.csv'):
    try:
        df = pd.read_csv(file)
        if 'Genres' in df.columns:
            for val in df['Genres'].dropna():
                for g in str(val).split(','):
                    g = g.strip()
                    if g: genres.add(g)
    except: continue

print(sorted(genres, key=str.lower))
print(f"\nTotal: {len(genres)}")
```
