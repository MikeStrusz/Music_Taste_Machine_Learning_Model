```python
import pandas as pd
import requests
from concurrent.futures import ThreadPoolExecutor
import random

covers_df = pd.read_csv('data/nmf_album_covers.csv')
if 'Album Name' in covers_df.columns and 'Album' not in covers_df.columns:
    covers_df = covers_df.rename(columns={'Album Name': 'Album'})

urls = covers_df[covers_df['Album Art'].notna() & covers_df['Album Art'].str.startswith('http')]['Album Art'].tolist()
sample = random.sample(urls, min(50, len(urls)))

def check_url(url):
    try:
        r = requests.head(url, timeout=5, headers={'User-Agent': 'Mozilla/5.0'})
        return 'alive' if r.status_code == 200 else f'dead ({r.status_code})'
    except:
        return 'dead (timeout/error)'

with ThreadPoolExecutor(max_workers=10) as ex:
    results = list(ex.map(check_url, sample))

from collections import Counter
print(Counter(results))
```

```python
import os
import pandas as pd

covers_df = pd.read_csv('data/nmf_album_covers.csv')
if 'Album Name' in covers_df.columns and 'Album' not in covers_df.columns:
    covers_df = covers_df.rename(columns={'Album Name': 'Album'})

cover_files = set(os.listdir('covers'))

def is_local(val):
    if pd.isna(val): return False
    val = str(val).replace('\\', '/')
    filename = val.split('/')[-1]
    return filename in cover_files

covers_df['has_local'] = covers_df['Album Art'].apply(is_local)

print(f"Already have local file: {covers_df['has_local'].sum()}")
print(f"URL only (need downloading): {(~covers_df['has_local'] & covers_df['Album Art'].notna()).sum()}")
print(f"Missing entirely: {covers_df['Album Art'].isna().sum()}")

# Show the ones that need downloading
need_download = covers_df[~covers_df['has_local'] & covers_df['Album Art'].notna()]
print("\nNeed downloading:")
print(need_download[['Artist', 'Album', 'Album Art']].to_string())
```

```python
import os
import pandas as pd
import re

covers_df = pd.read_csv('data/nmf_album_covers.csv')
if 'Album Name' in covers_df.columns and 'Album' not in covers_df.columns:
    covers_df = covers_df.rename(columns={'Album Name': 'Album'})

cover_files = set(os.listdir('covers'))

def url_to_local(val):
    if pd.isna(val): return None
    match = re.search(r'([a-f0-9]{32})\.(jpg|png|webp)', str(val))
    if match:
        filename = f"{match.group(1)}.{match.group(2)}"
        if filename in cover_files:
            return f"covers/{filename}"
    return None

covers_df['local_path'] = covers_df['Album Art'].apply(url_to_local)

print(f"URLs that match a local file: {covers_df['local_path'].notna().sum()}")
print(f"URLs with no local match: {covers_df['local_path'].isna().sum()}")
```

```python
import hashlib, os, pandas as pd

# Load covers CSV
covers_df = pd.read_csv('data/nmf_album_covers.csv')
if 'Artist Name(s)' in covers_df.columns:
    covers_df = covers_df.rename(columns={'Artist Name(s)': 'Artist'})
if 'Album Name' in covers_df.columns and 'Album' not in covers_df.columns:
    covers_df = covers_df.rename(columns={'Album Name': 'Album'})

cover_files = set(os.listdir('covers'))

def compute_hash_filename(artist, album):
    key = f"{str(artist).strip().lower()}_{str(album).strip().lower()}"
    return hashlib.md5(key.encode()).hexdigest() + '.jpg'

# Test on first 10 rows
for _, row in covers_df.head(10).iterrows():
    fn = compute_hash_filename(row['Artist'], row['Album'])
    found = fn in cover_files
    print(f"{'✅' if found else '❌'} {row['Artist']} - {row['Album'][:30]}: {fn}")
```

```python
matched = 0
not_found = 0

for idx, row in covers_df.iterrows():
    fn = compute_hash_filename(row['Artist'], row['Album'])
    if fn in cover_files:
        covers_df.at[idx, 'Album Art'] = f"covers/{fn}"
        matched += 1
    else:
        not_found += 1

print(f"Matched: {matched}, No local file: {not_found}")
covers_df.to_csv('data/nmf_album_covers.csv', index=False)
print("Saved!")
```
