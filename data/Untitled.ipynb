```python
quiet_instrumentals = (
    df[
        (df["Instrumentalness"] >= 0.68) &
        (df["Instrumentalness"] <= 0.72) &
        (df["Duration (ms)"] >= 2 * 60 * 1000) &
        (df["Duration (ms)"] <= 5 * 60 * 1000)
    ]
    .sort_values("Loudness")  # more negative = quieter
    .head(10)
)

quiet_instrumentals[
    ["Track Name", "Artist Name(s)", "Instrumentalness", "Loudness", "Duration (ms)"]
]
```

```python
# Filter pop tracks with low tempo (≤100 BPM)
pop_slow = df[
    df["Genres"].str.contains("pop", case=False, na=False) &
    (df["Tempo"] <= 100)
]

# Show the relevant stats
pop_slow[["Track Name", "Artist Name(s)", "Genres", "Tempo", "Duration (ms)"]]
```

```python
pop_slow_sorted = pop_slow.sort_values("Tempo")  # ascending = slowest first

pop_slow_sorted[["Track Name", "Artist Name(s)", "Genres", "Tempo", "Duration (ms)"]]
```

```python
# Load pandas if not already
import pandas as pd

# Load your mid.csv playlist
mid_df = pd.read_csv("mid.csv")

# Quick check of columns
mid_df.columns, mid_df.head(2)
```

```python
# Filter pop tracks with low tempo (≤100 BPM)
pop_slow_mid = mid_df[
    mid_df["Genres"].str.contains("pop", case=False, na=False) &
    (mid_df["Tempo"] <= 100)
]

# Sort by slowest tempo first
pop_slow_mid_sorted = pop_slow_mid.sort_values("Tempo")

# Show relevant columns
pop_slow_mid_sorted[["Track Name", "Artist Name(s)", "Genres", "Tempo", "Duration (ms)"]].head(10)
```

```python
# Keep only the desired pop subgenres
desired_genres = ["indie pop", "bedroom pop", "pop"]

# Filter
filtered_pop_mid = mid_df[
    mid_df["Genres"].str.contains("|".join(desired_genres), case=False, na=False) &  # keep only desired
    ~mid_df["Genres"].str.contains("baroque pop", case=False, na=False) &           # remove baroque pop
    ~mid_df["Artist Name(s)"].str.contains("Beirut", case=False, na=False)          # remove Beirut
]

# Sort by slowest tempo first
filtered_pop_mid_sorted = filtered_pop_mid.sort_values("Tempo")

# Show relevant columns
filtered_pop_mid_sorted[["Track Name", "Artist Name(s)", "Genres", "Tempo", "Duration (ms)"]].head(30)
```
