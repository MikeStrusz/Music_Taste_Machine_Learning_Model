```python
# Load pandas if not already
import pandas as pd

# Load your mid.csv playlist
mid_df = pd.read_csv("mid.csv")

# Quick check of columns
mid_df.columns, mid_df.head(2)
```

```python
# Filter pop tracks with low tempo (≤100 BPM)
pop_slow_mid = mid_df[
    mid_df["Genres"].str.contains("pop", case=False, na=False) &
    (mid_df["Tempo"] <= 100)
]

# Sort by slowest tempo first
pop_slow_mid_sorted = pop_slow_mid.sort_values("Tempo")

# Show relevant columns
pop_slow_mid_sorted[["Track Name", "Artist Name(s)", "Genres", "Tempo", "Duration (ms)"]].head(10)
```

```python
# Keep only the desired pop subgenres
desired_genres = ["indie pop", "bedroom pop", "pop"]

# Filter
filtered_pop_mid = mid_df[
    mid_df["Genres"].str.contains("|".join(desired_genres), case=False, na=False) &  # keep only desired
    ~mid_df["Genres"].str.contains("baroque pop", case=False, na=False) &           # remove baroque pop
    ~mid_df["Artist Name(s)"].str.contains("Beirut", case=False, na=False)          # remove Beirut
]

# Sort by slowest tempo first
filtered_pop_mid_sorted = filtered_pop_mid.sort_values("Tempo")

# Show relevant columns
filtered_pop_mid_sorted[["Track Name", "Artist Name(s)", "Genres", "Tempo", "Duration (ms)"]].head(30)
```

```python
import pandas as pd, requests, os
from time import sleep

covers_dir = 'covers'
os.makedirs(covers_dir, exist_ok=True)

covers_df = pd.read_csv('data/nmf_album_covers.csv')

downloaded = 0
skipped = 0
failed = 0

for _, row in covers_df.iterrows():
    artist = str(row['Artist']).strip()
    album = str(row['Album Name']).strip()
    url = str(row['Album Art']).strip() if pd.notna(row['Album Art']) else ''
    
    if not url:
        skipped += 1
        continue
    
    # Build a safe filename
    safe_name = f"{artist} - {album}".replace('/', '_').replace('\\', '_').replace(':', '_').replace('*', '_').replace('?', '_').replace('"', '_').replace('<', '_').replace('>', '_').replace('|', '_')
    local_path = os.path.join(covers_dir, f"{safe_name}.jpg")
    
    if os.path.exists(local_path):
        skipped += 1
        continue
    
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        with open(local_path, 'wb') as f:
            f.write(response.content)
        downloaded += 1
        sleep(0.05)
    except Exception as e:
        print(f"  ❌ {artist} - {album}: {e}")
        failed += 1

print(f"\n✅ Done! Downloaded: {downloaded} | Skipped: {skipped} | Failed: {failed}")
print(f"Covers saved to: {covers_dir}/")
```
