```python
import pandas as pd

df = pd.read_csv('data/2026_training_complete_with_features.csv')

# Check how many Top 100 albums have missing Genres or Record Label
top100 = df[df['source_type'] == 'top_100_ranked'].drop_duplicates(subset=['Album Name'])

print(f"Total Top 100 albums: {len(top100)}")
print(f"Missing Genres: {top100['Genres'].isna().sum()}")
print(f"Missing Record Label: {top100['Record Label'].isna().sum()}")
print()
print("Sample of what IS populated:")
print(top100[top100['Genres'].notna()][['Album Name', 'Artist Name(s)', 'Genres', 'Record Label']].head(5).to_string())
```

```python
import pandas as pd

df = pd.read_csv('data/2026_training_complete_with_features.csv')
top100_missing = df[
    (df['source_type'] == 'top_100_ranked') & 
    (df['Genres'].isna())
].drop_duplicates(subset=['Album Name'])

print(f"Albums to backfill: {len(top100_missing)}")
print()
print("Sample of what needs genres:")
print(top100_missing[['Album Name', 'Artist Name(s)', 'Record Label']].head(10).to_string())
```
