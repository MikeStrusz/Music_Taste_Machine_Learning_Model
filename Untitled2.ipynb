```python
# One‑time vault cover fetch (2020–2025)
import pandas as pd
import os
import time
import requests
from fuzzywuzzy import fuzz
import hashlib

# --- Re‑use the album art fetcher from the notebook ---
def get_album_art(artist: str, album: str) -> dict:
    """Improved album art fetcher with fuzzy matching fallback and status logging."""
    try:
        # 1. Try exact search first
        url = f"https://itunes.apple.com/search?term={artist} {album}&entity=album"
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        results = data.get('results', [])
        
        # 2. Fallback: Search by Artist only and use fuzzy matching for the album
        if not results:
            url = f"https://itunes.apple.com/search?term={artist}&entity=album"
            response = requests.get(url)
            data = response.json()
            results = data.get('results', [])
            
            if results:
                best_match = None
                highest_score = 0
                for res in results:
                    score = fuzz.ratio(str(album).lower(), str(res.get('collectionName', '')).lower())
                    if score > highest_score:
                        highest_score = score
                        best_match = res
                
                if highest_score > 60:
                    results = [best_match]
                    print(f"  🔍 Fuzzy Match Found: '{artist} - {best_match.get('collectionName')}' (Score: {highest_score})")
                else:
                    results = []

        if results:
            album_art_url = results[0].get('artworkUrl100', '').replace("100x100", "600x600")
            # Try to download to local covers folder
            try:
                covers_dir = 'covers'
                os.makedirs(covers_dir, exist_ok=True)
                key = f"{str(artist).strip().lower()}_{str(album).strip().lower()}"
                local_filename = hashlib.md5(key.encode()).hexdigest() + '.jpg'
                local_path = os.path.join(covers_dir, local_filename)
                if not os.path.exists(local_path):
                    img_response = requests.get(album_art_url, timeout=10)
                    img_response.raise_for_status()
                    with open(local_path, 'wb') as f:
                        f.write(img_response.content)
                album_art = local_path  # store local path
            except Exception as e:
                print(f"  ⚠️ Download failed, storing URL instead: {e}")
                album_art = album_art_url  # fall back to URL if download fails
            return {'Artist': artist, 'Album Name': album, 'Album Art': album_art, 'status': 'success'}
        else:
            print(f"  ❌ No Results: {artist} - {album}")
            
    except Exception as e:
        print(f"  ⚠️ Error fetching {artist}: {e}")
        return {'Artist': artist, 'Album Name': album, 'Album Art': None, 'status': f'error: {str(e)}'}
    
    return {'Artist': artist, 'Album Name': album, 'Album Art': None, 'status': 'no_results'}

# --- Load vault albums and existing covers ---
vault_file = 'data/vault_clean.csv'
covers_file = 'data/nmf_album_covers.csv'

if not os.path.exists(vault_file):
    raise FileNotFoundError("vault_clean.csv not found. Please run the vault data prep first.")

vault_df = pd.read_csv(vault_file)
# Use artist/album columns from vault_clean
# Ensure consistent column names
if 'Artist' not in vault_df.columns:
    # fallback: maybe it's 'Artist Name' or something
    print("Warning: 'Artist' column not found. Check column names.")
    print("Columns:", vault_df.columns.tolist())
    # You may need to adjust the column names here
    # For now, assume 'Artist' and 'Album' exist

# Load existing covers
existing_covers = pd.DataFrame()
if os.path.exists(covers_file):
    existing_covers = pd.read_csv(covers_file)
    print(f"Loaded {len(existing_covers)} existing covers.")
else:
    print("No existing covers file found. Will create new one.")

# Build set of already covered albums (Artist, Album Name)
covered = set()
if not existing_covers.empty:
    covered = set(zip(existing_covers['Artist'], existing_covers['Album Name']))

# Get all unique artist/album pairs from vault (skip duplicates)
unique_albums = vault_df[['Artist', 'Album']].drop_duplicates().copy()
unique_albums.columns = ['Artist', 'Album Name']

# Only fetch those not already covered
to_fetch = unique_albums[~unique_albums.apply(lambda r: (r['Artist'], r['Album Name']) in covered, axis=1)]
print(f"Total unique vault albums: {len(unique_albums)}")
print(f"Already covered: {len(covered)}")
print(f"To fetch: {len(to_fetch)}")

# Fetch and save
new_covers = []
for idx, row in to_fetch.iterrows():
    artist = row['Artist']
    album = row['Album Name']
    print(f"Fetching {artist} - {album}")
    result = get_album_art(artist, album)
    if result.get('Album Art'):
        new_covers.append(result)
    time.sleep(0.2)  # be polite to the API

if new_covers:
    new_covers_df = pd.DataFrame(new_covers)[['Artist', 'Album Name', 'Album Art']]
    # Append to existing covers file
    updated_covers = pd.concat([existing_covers, new_covers_df], ignore_index=True)
    updated_covers.to_csv(covers_file, index=False)
    print(f"✅ Added {len(new_covers)} new covers. Total covers now: {len(updated_covers)}")
else:
    print("No new covers fetched.")
```

```python
import os
from PIL import Image

covers_dir = 'covers'
max_size = 300

if not os.path.exists(covers_dir):
    print("Covers folder not found.")
else:
    resized_count = 0
    for filename in os.listdir(covers_dir):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            filepath = os.path.join(covers_dir, filename)
            try:
                with Image.open(filepath) as img:
                    # Only resize if larger than max_size
                    if img.width > max_size or img.height > max_size:
                        img.thumbnail((max_size, max_size), Image.Resampling.LANCZOS)
                        img.save(filepath, optimize=True, quality=85)
                        resized_count += 1
            except Exception as e:
                print(f"Error processing {filename}: {e}")
    print(f"✅ Resized {resized_count} images to max {max_size}px.")
```
